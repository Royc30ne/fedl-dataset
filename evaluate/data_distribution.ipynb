```python
# Dependency
import os
import numpy as np
from collections import defaultdict
```

```python
# Load Dataset
fed_dataset = 'emnist_digits_2000_iid'
path = '../fedl_data/' + fed_dataset + '/'

# Load Data
def load_federated_data(path, prefix):
    # match prefix_0_x.npy, prefix_1_x.npy, ... and prefix_0_y.npy, prefix_1_y.npy, ...
    federated_data = defaultdict(lambda: {'x': [], 'y': []})
    i = 0
    while True:
        x_file = path + prefix + '_' + str(i) + '_x.npy'
        y_file = path + prefix + '_' + str(i) + '_y.npy'
        if os.path.exists(x_file) and os.path.exists(y_file):
            x_list = np.load(x_file)
            y_list = np.load(y_file)
            federated_data[prefix + '_' + str(i)]['x'] = x_list
            federated_data[prefix + '_' + str(i)]['y'] = y_list
            print('client: ', i, 'x: ', x_list.shape, 'y: ', y_list.shape)
            i += 1
        else:
            print('loading finished, total clients: ', i)
            break
    return federated_data

train_path = path + 'train/'
test_path = path + 'test/'
train_data = load_federated_data(train_path, 'client')
print('train clients: ', len(train_data))
```

```python
print(train_data['client_0']['y'])
```

```python
# Get the label distribution for each client
label_distribution = {}
for client in train_data:
    y_list = train_data[client]['y']
    label_distribution[client] = {}
    for y in y_list:
        if y in label_distribution[client]:
            label_distribution[client][y] += 1
        else:
            label_distribution[client][y] = 1
print('total clients: ', len(label_distribution))
# Calculate the average label distribution
label_distribution_avg = {}
for client in label_distribution:
    for label in label_distribution[client]:
        if label in label_distribution_avg:
            label_distribution_avg[label] += label_distribution[client][label]
        else:
            label_distribution_avg[label] = label_distribution[client][label]
            
for i in label_distribution:
    print(i, label_distribution[i])
```
